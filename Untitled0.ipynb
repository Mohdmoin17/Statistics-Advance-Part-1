Part 1: Theoretical Questions & Answers
1. What is a random variable in probability theory?
A random variable is a numerical quantity whose value depends on the outcomes of a random phenomenon. It's a function that assigns a real number to each outcome in a sample space. (e.g., the number of heads in 10 coin tosses).

2. What are the types of random variables?

Discrete Random Variables: Can take on a finite or countably infinite set of values (e.g., number of students in a class).

Continuous Random Variables: Can take on any value in an interval on the number line (e.g., height of a person, time it takes to run a race).

3. What is the difference between discrete and continuous distributions?
A discrete distribution describes the probabilities of a discrete random variable, defined by a Probability Mass Function (PMF). A continuous distribution describes the probabilities of a continuous random variable, defined by a Probability Density Function (PDF), where probabilities are calculated for intervals and the probability of a single point is zero.

4. What are probability distribution functions (PDF)?
This term can be ambiguous.

For continuous variables, the Probability Density Function (PDF) is a function whose area under the curve between two points defines the probability that the variable lies in that interval.

For discrete variables, the Probability Mass Function (PMF) is a function that gives the probability that a discrete random variable is exactly equal to some value.

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
The Cumulative Distribution Function (CDF), F(x), gives the probability that a random variable X is less than or equal to a value x (i.e., F(x) = P(X ≤ x)). The PDF/PMF gives the probability at a specific point (for discrete) or the density at a point (for continuous). The CDF is the integral of the PDF for continuous variables, and the sum of the PMF for discrete variables.

6. What is a discrete uniform distribution?
A distribution where a finite number of values are equally likely to occur. A classic example is a fair die roll where each face (1 through 6) has a probability of 1/6.

7. What are the key properties of a Bernoulli distribution?
It's a discrete distribution for a single trial with only two possible outcomes: "success" (1) with probability p, and "failure" (0) with probability q = 1 - p.

8. What is the binomial distribution, and how is it used in probability?
The binomial distribution models the number of successes k in a fixed number n of independent Bernoulli trials, each with the same success probability p. It's used to calculate probabilities like "the chance of getting exactly 5 heads in 10 coin flips."

9. What is the Poisson distribution and where is it applied?
A discrete distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space, if these events occur with a known constant mean rate and independently of the time since the last event. It's applied in fields like telecommunications (number of phone calls per hour), finance (number of trades per minute), and biology (number of mutations on a strand of DNA).

10. What is a continuous uniform distribution?
A distribution where all intervals of the same length have an equal probability. It's defined by two parameters, a (minimum) and b (maximum). The PDF is a constant horizontal line between a and b.

11. What are the characteristics of a normal distribution?

Bell-shaped and symmetric about its mean.

Mean, median, and mode are all equal.

Defined by two parameters: mean (μ, determines location) and standard deviation (σ, determines spread).

The empirical rule (68-95-99.7 rule) applies.

12. What is the standard normal distribution, and why is it important?
It is a special case of the normal distribution with a mean (μ) of 0 and a standard deviation (σ) of 1. It's important because any normal distribution can be converted to a standard normal distribution using Z-scores, allowing us to use a single standard table (Z-table) to calculate probabilities.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
The CLT states that the sampling distribution of the sample mean will approach a normal distribution as the sample size n gets larger, regardless of the shape of the population distribution. This is critical because it allows us to make inferences about population parameters (like the mean) using normal probability theory, even if the underlying population is not normal.

14. How does the Central Limit Theorem relate to the normal distribution?
The CLT explains why the normal distribution is so prevalent. It guarantees that for large enough sample sizes, the distribution of sample means will be approximately normal, making the normal distribution the foundation for much of inferential statistics (e.g., confidence intervals, hypothesis tests).

15. What is the application of Z statistics in hypothesis testing?
Z-statistics (or Z-tests) are used to test hypotheses about population means when the population standard deviation is known and the sample size is large (or the population is normal). The calculated Z-score tells us how many standard errors the sample mean is from the hypothesized population mean, which we then use to calculate a p-value.

16. How do you calculate a Z-score, and what does it represent?
The Z-score for a data point is calculated as: Z = (X - μ) / σ
It represents the number of standard deviations a data point X is from the population mean μ. It's a measure of relative standing.

17. What are point estimates and interval estimates in statistics?

Point Estimate: A single value used to estimate a population parameter (e.g., using the sample mean x̄ to estimate the population mean μ).

Interval Estimate: A range of values (an interval) within which the population parameter is estimated to lie, with a certain level of confidence (e.g., a 95% confidence interval).

18. What is the significance of confidence intervals in statistical analysis?
Confidence intervals provide a range of plausible values for an unknown population parameter. The confidence level (e.g., 95%) expresses the degree of certainty that the interval contains the true parameter. It gives more information than a single point estimate by also conveying the precision of the estimate.

19. What is the relationship between a Z-score and a confidence interval?
The construction of a confidence interval for a mean (when σ is known) relies directly on the Z-score from the standard normal distribution. The formula for a CI is: x̄ ± Z* * (σ/√n), where Z* is the critical Z-value corresponding to the desired confidence level (e.g., 1.96 for 95% confidence).

20. How are Z-scores used to compare different distributions?
By converting different datasets to Z-scores (a process called standardization), you put them on the same scale (mean=0, std=1). This allows for direct comparison of values from different distributions, answering questions like "Is an IQ of 130 more exceptional than a SAT score of 700?"

21. What are the assumptions for applying the Central Limit Theorem?
The key assumptions are:

Random Sampling: The data must be a random sample from the population.

Independence: Sample observations are independent.

Sample Size: A "large enough" sample size (n). Often n >= 30 is considered sufficient, but it depends on the skewness of the original population.

22. What is the concept of expected value in a probability distribution?
The expected value is the long-run average value of repetitions of the experiment it represents. It's a weighted average of all possible values, weighted by their probabilities. For a discrete variable, it's calculated as E[X] = Σ x * P(X=x).

23. How does a probability distribution relate to the expected outcome of a random variable?
The probability distribution defines the expected outcome. The expected value is the central, average value you would predict from the random variable after many, many trials, based on its underlying probability distribution.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1. Generate a random variable and display its value.
print("1. Random variable (from standard normal):", np.random.randn())

# 2. Generate a discrete uniform distribution (e.g., die roll) and plot PMF.
print("\n2. & 6. Discrete Uniform Distribution (Die Roll)")
num_rolls = 1000
rolls = np.random.randint(1, 7, size=num_rolls) # 1000 die rolls (1-6)
values, counts = np.unique(rolls, return_counts=True)
pmf = counts / num_rolls

# Plot PMF
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.stem(values, pmf, basefmt=" ")
plt.title('PMF of Discrete Uniform (Die Roll)')
plt.xlabel('Value')
plt.ylabel('Probability')

# 6. Calculate and plot CDF for the same distribution
cdf = np.cumsum(pmf)
plt.subplot(1, 2, 2)
plt.step(values, cdf, where='mid')
plt.title('CDF of Discrete Uniform (Die Roll)')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.tight_layout()
plt.show()

# 3. PDF of a Bernoulli distribution
def bernoulli_pdf(p, k):
    """Calculates P(X=k) for a Bernoulli(p) distribution."""
    return p if k == 1 else (1 - p)

p_val = 0.7
print(f"\n3. Bernoulli PDF: P(X=1) = {bernoulli_pdf(p_val, 1):.2f}, P(X=0) = {bernoulli_pdf(p_val, 0):.2f}")

# 4. Simulate and plot a Binomial distribution (n=10, p=0.5)
print("\n4. Binomial Distribution")
n, p = 10, 0.5
binomial_data = np.random.binomial(n, p, 1000)
plt.hist(binomial_data, bins=np.arange(-0.5, n+1.5, 1), density=True, edgecolor='black', alpha=0.7)
plt.title('Histogram of Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()

# 5. Create and visualize a Poisson distribution
print("\n5. Poisson Distribution")
lambda_param = 4 # mean rate
poisson_data = np.random.poisson(lambda_param, 1000)
values, counts = np.unique(poisson_data, return_counts=True)
pmf_poisson = counts / len(poisson_data)

plt.stem(values, pmf_poisson, basefmt=" ")
plt.title(f'PMF of Poisson Distribution (λ = {lambda_param})')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.show()

# 7. Generate and visualize a Continuous Uniform distribution
print("\n7. Continuous Uniform Distribution")
low, high = 5, 15
continuous_uniform = np.random.uniform(low, high, 1000)
plt.hist(continuous_uniform, density=True, bins=30, edgecolor='black', alpha=0.7)
plt.title(f'Continuous Uniform Distribution [{low}, {high}]')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

# 8. Simulate and plot data from a normal distribution
print("\n8. Normal Distribution")
mu, sigma = 0, 1
normal_data = np.random.normal(mu, sigma, 1000)
plt.hist(normal_data, bins=30, density=True, edgecolor='black', alpha=0.7)
plt.title(f'Histogram of Normal Data (μ={mu}, σ={sigma})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

# 9. & 27. Calculate Z-scores and plot them (Standardization)
print("\n9. & 27. Z-Score Calculation and Visualization")
# Generate some sample data that is not standard normal
original_data = np.random.exponential(scale=2, size=500)
z_scores = (original_data - original_data.mean()) / original_data.std()

# Plot original vs. standardized
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(original_data, bins=30, edgecolor='black', alpha=0.7)
plt.title('Original Data (Exponential)')

plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, edgecolor='black', alpha=0.7)
plt.title('Standardized Data (Z-Scores)')
plt.xlabel('Z-score')
plt.tight_layout()
plt.show()

# 15. Simulate CLT: Sample from a non-normal distribution (Exponential)
print("\n15. Central Limit Theorem Simulation")
pop_size = 10000
sample_size = 30
num_samples = 1000

# Create a non-normal population
population = np.random.exponential(scale=1, size=pop_size)
# Take many samples and calculate their means
sample_means = [np.random.choice(population, size=sample_size, replace=False).mean() for _ in range(num_samples)]

# Plot population distribution and sampling distribution
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(population, bins=50, density=True, edgecolor='black', alpha=0.7)
plt.title('Population Distribution (Exponential)')

plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, density=True, edgecolor='black', alpha=0.7)
# Overlay a normal PDF with the same mean and std
x = np.linspace(min(sample_means), max(sample_means), 100)
plt.plot(x, stats.norm.pdf(x, np.mean(sample_means), np.std(sample_means)), 'r-', lw=2)
plt.title('Sampling Distribution of the Mean')
plt.suptitle('Demonstration of the Central Limit Theorem')
plt.tight_layout()
plt.show()

# 16. Calculate and plot the standard normal distribution
print("\n16. Standard Normal Distribution")
x = np.linspace(-4, 4, 100)
pdf = stats.norm.pdf(x, 0, 1) # PDF with mean=0, std=1
cdf = stats.norm.cdf(x, 0, 1) # CDF with mean=0, std=1

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(x, pdf)
plt.title('Standard Normal PDF')
plt.xlabel('x')
plt.ylabel('Density')

plt.subplot(1, 2, 2)
plt.plot(x, cdf)
plt.title('Standard Normal CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.tight_layout()
plt.show()

# 18. Calculate Z-score for a given data point
print("\n18. Calculate Z-score for a data point")
data = np.random.normal(50, 10, 100) # Fake dataset: mean=50, std=10
point_of_interest = 65
z_score_manual = (point_of_interest - np.mean(data)) / np.std(data, ddof=1) # Use sample std
print(f"Z-score for {point_of_interest}: {z_score_manual:.2f}")

# 20. & 21. Create a confidence interval for a mean (assuming population std is known)
print("\n20. & 21. Confidence Interval for Mean")
# Let's use the `data` from above. Assume we *know* the population std is 10.
confidence_level = 0.95
z_critical = stats.norm.ppf((1 + confidence_level) / 2) # Get Z* for 95% CI, e.g., ~1.96
margin_of_error = z_critical * (10 / np.sqrt(len(data))) # (pop_std / sqrt(n))
ci_low = np.mean(data) - margin_of_error
ci_high = np.mean(data) + margin_of_error
print(f"95% Confidence Interval: ({ci_low:.2f}, {ci_high:.2f})")
# Interpretation: We are 95% confident that the true population mean lies between ci_low and ci_high.

# 22. Plot PDF of a normal distribution
print("\n22. PDF of a Normal Distribution")
mu_plot, sigma_plot = 100, 15
x = np.linspace(mu_plot - 4*sigma_plot, mu_plot + 4*sigma_plot, 100)
pdf = stats.norm.pdf(x, mu_plot, sigma_plot)
plt.plot(x, pdf)
plt.title(f'Normal PDF (μ={mu_plot}, σ={sigma_plot})')
plt.xlabel('x')
plt.ylabel('Density')
plt.show()

# 23. Calculate and interpret CDF of a Poisson distribution
print("\n23. CDF of a Poisson Distribution")
lambda_pois = 3
k = 2
cdf_value = stats.poisson.cdf(k, lambda_pois) # P(X <= 2)
print(f"CDF at k={k} for Poisson(λ={lambda_pois}): P(X <= {k}) = {cdf_value:.3f}")
# Interpretation: The probability of observing 2 or fewer events is ~0.423.

# 24. Simulate continuous uniform and calculate expected value
print("\n24. Continuous Uniform Expected Value")
a, b = 10, 20
data_uniform = np.random.uniform(a, b, 10000)
expected_value_simulated = np.mean(data_uniform)
expected_value_theoretical = (a + b) / 2
print(f"Theoretical E[X]: {expected_value_theoretical}")
print(f"Simulated E[X] (from 10,000 samples): {expected_value_simulated:.2f}")

# 28. Calculate skewness and kurtosis of a normal dataset
print("\n28. Skewness and Kurtosis")
normal_data = np.random.normal(0, 1, 1000)
skewness = stats.skew(normal_data)
kurtosis = stats.kurtosis(normal_data) # Fisher's definition, excess kurtosis (0 for normal)
print(f"Skewness: {skewness:.3f} (should be near 0)")
print(f"Excess Kurtosis: {kurtosis:.3f} (should be near 0)")
```
